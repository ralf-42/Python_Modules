# 5 | Multi-Speaker Audio ‚≠ê
---

<p><font color='black' size="5">
2-Personen Dialog
</font></p>

```python
# Beispiel-Dialog-Datei erstellen
dialog_text = """Sprecher 1: Guten Tag und herzlich willkommen zu unserem Interview √ºber k√ºnstliche Intelligenz!
Sprecher 2: Vielen Dank f√ºr die Einladung! Ich freue mich, heute hier zu sein.
Sprecher 1: Erz√§hlen Sie uns doch bitte, woran Sie aktuell arbeiten.
Sprecher 2: Ich besch√§ftige mich haupts√§chlich mit generativen KI-Modellen, insbesondere mit Text-to-Speech und Speech-to-Text Systemen.
Sprecher 1: Das klingt sehr spannend! Wie sehen Sie die Zukunft dieser Technologie?
Sprecher 2: Ich denke, wir stehen erst am Anfang. Die Modelle werden immer nat√ºrlicher klingen und vielf√§ltiger einsetzbar sein.
Sprecher 1: Vielen Dank f√ºr diese interessanten Einblicke!
Sprecher 2: Gerne! Es war mir eine Freude.
"""

with open('demo_dialog_2speaker.txt', 'w', encoding='utf-8') as f:
    f.write(dialog_text)

print("‚úÖ Dialog-Datei erstellt: demo_dialog_2speaker.txt")
```

```python
# Stimmen f√ºr Sprecher definieren
speaker_voices = {
    1: 'alloy',   # Interviewer (neutral)
    2: 'nova',    # Expertin (weiblich)
}

# Audio erstellen
output_file = create_multi_speaker_audio(
    text_file='demo_dialog_2speaker.txt',
    speaker_voices=speaker_voices,
    output_file='demo_dialog_2speaker.mp3',
    config=config,
    pause_duration=600  # 600ms Pause zwischen Sprechern
)

display(Markdown("## üéôÔ∏è 2-Personen Interview"))
display(Markdown("---"))
print(f"‚úÖ Multi-Speaker Audio erstellt: {output_file}")
print()
display(Audio(output_file, autoplay=False))
```

# 6 | Speech-to-Text (STT)
---

<p><font color='black' size="5">
Einfache Transkription
</font></p>

```python
# Transkribiere eine der erstellten Audiodateien
audio_file = 'demo_simple.mp3'

transcript = transcribe_audio(audio_file, config)

display(Markdown("## üìÑ Transkription"))
display(Markdown("---"))
display(Markdown(f"**Datei:** `{audio_file}`"))
display(Markdown(f"**Text:** {transcript}"))
```

<p><font color='black' size="5">
Transkription mit Zeitstempeln
</font></p>

```python
# Transkription mit Zeitstempeln
audio_file = 'demo_dialog_2speaker.mp3'

segments = transcribe_with_timestamps(audio_file, config)

display(Markdown("## üìÑ Transkription mit Zeitstempeln"))
display(Markdown("---"))
display(Markdown(f"**Datei:** `{audio_file}`"))
display(Markdown(""))

for seg in segments:
    display(Markdown(f"`{seg['start']:.2f}s - {seg['end']:.2f}s:` {seg['text']}"))
```

```python
# Podcast mit expliziten Intro/Outro
intro = "Willkommen zu unserem Podcast √ºber k√ºnstliche Intelligenz und ihre Anwendungen!"

main_content = """
In der heutigen Folge sprechen wir √ºber Text-to-Speech Systeme.
Diese Systeme haben in den letzten Jahren enorme Fortschritte gemacht.
Die Sprachqualit√§t ist mittlerweile kaum noch von echter menschlicher Sprache zu unterscheiden.
Besonders interessant ist die M√∂glichkeit, verschiedene Stimmen f√ºr unterschiedliche Sprecher zu nutzen.
"""

outro = "Vielen Dank f√ºrs Zuh√∂ren! In der n√§chsten Folge sprechen wir √ºber Speech-to-Text. Bis bald!"

output_file = create_podcast(
    intro_text=intro,
    main_content=main_content,
    outro_text=outro,
    output_file='demo_podcast_standard.mp3',
    config=config,
    voices={
        'intro': 'nova',
        'main': 'alloy',
        'outro': 'nova'
    },
    pause_duration=1200
)

display(Markdown("## üéß Standard-Podcast"))
display(Markdown("---"))
print(f"‚úÖ Podcast erstellt: {output_file}")
print()
display(Audio(output_file, autoplay=False))
```
